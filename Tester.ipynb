```python
import cv2 
import os 
import numpy as np
import import_ipynb
import facialRecognition as fr
```

```python
def cv_imread(file_path):
    cv_img = cv2.imdecode(np.fromfile(file_path, dtype=np.uint8), -1)
    return cv_img

file_path = 'C:/Users/Asus/Documents/Kerja Praktik/FaceRecognition/dataset/2/ester31.jpg'
test_img = cv_imread(file_path)
print(test_img)
faces_detected, gray_img = fr.faceDetection(test_img)
print("face detected : ", faces_detected)

for (x,y,w,h) in faces_detected:
    cv2.rectangle(test_img, (x,y), (x+w, y+h), (255,0,0), thickness= 5)
    
faces, faceID = fr.imageLabelling('C:/Users/Asus/Documents/Kerja Praktik/FaceRecognition/dataset/')
face_recog = fr.train_model(faces, faceID)
face_recog.save('hasil_training.yml')

# face_recog = cv2.face.LBPHFaceRecognizer_create()
# face_recog.read('C:/Users/Asus/Documents/Kerja Praktik/FaceRecognition/trainRes.h5')

person_name = {1:"Irene Panjaitan", 2:"Ester Marbun", 3:"Joel Marpaung"}

for face in faces_detected:
    (x,y,w,h)= face
    img_gray = gray_img[y:y+h, x:x+h]
    label, confidence = face_recog.predict(img_gray)
    print("confidence:", confidence)
    print("label:", label)
    
    fr.make_rectangle(test_img, face)
    predict_person = person_name[label]
    
    if (confidence>10):
        continue
    
    fr.show_name(test_img, predict_person, x,y)
    
resized_img = cv2.resize(test_img, (1000, 700))
cv2.imshow("Face Recognition", resized_img)
cv2.waitKey(0)

cv2.destroyAllWindows()
cv2.waitKey(1)
```
